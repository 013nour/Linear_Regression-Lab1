```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

X = 2 * np.random.rand(500,)
y = (4 + 3 * X + np.random.randn(500,)) + X ** 2

plt.scatter(X, y, marker='x', c='r',label='Actual Values')

plt.title("Closed Form Disadvantage")

plt.ylabel('Y Target')
plt.xlabel('X Feature')
plt.legend()

plt.show();
```

```python
len(X)
```

```python
Xmean = X.mean()
Ymean = y.mean()

XSTD = X.std()
YSTD = y.std()

#R Corri Coffi
Tot = 0
for i /in range(len(X)):
    Tot += ((X[i]-Xmean)/XSTD)*((y[i]-Ymean)/YSTD)
    
R = (1/(len(X)-1)) * Tot

#find the tangent line equation y=mx+b
m = R*(YSTD/XSTD)
b = Ymean - (m*Xmean)

print(m)
print(b)

# y=5.046X + 3.19
y_hat = []
for i in range(len(X)):
    j = m * X[i] + b
    y_hat.append(j)
    
#(w,b) found by gradient descent: (5.036, 3.2039)
```

```python
plt.plot(X, y_hat, c='b',label='Our Prediction')
plt.scatter(X, y, marker='x', c='r',label='Actual Values')
```
